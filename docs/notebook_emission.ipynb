```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Combine all experiment data into a single DataFrame (you can add more experiments as needed)
data_combined = {
    'timestamp': [
        '2024-10-02T00:17:11', '2024-10-02T00:37:29', '2024-10-02T00:10:17', 
        '2024-10-02T00:29:00'
    ],
    'project_name': ['Dog_breed_classification_model']*4,
    'experiment_type': [
        'mobilenet_batch_64', 'mobilenet_batch_32_rmsprop',
        'mobilenet_batch_32', 'mobilenet_batch_32_adam'
    ],
    'emissions': [
        0.00011436933774721845, 0.00020518576677416514,
        0.0002282911985339208, 0.00024680394776007835
    ]
}

# Creating the combined dataframe
df_combined = pd.DataFrame(data_combined)
```

```python
experiment_types = data_combined['experiment_type']
emissions = data_combined['emissions']

# Create the bar plot
fig, ax = plt.subplots()
bar_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Assign colors for each bar
bars = ax.bar(experiment_types, emissions, color=bar_colors)

# Add labels and title
ax.set_xlabel('Experiment Type')
ax.set_ylabel('Emissions')
ax.set_title('Emissions for Different Experiment Types')

# Display the plot
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
